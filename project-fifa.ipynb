# Step 1: Cleaning the dataset

```python
## we first displayed all the categorical columns and evaluated their relevance to the performance. 
## we then checked the nb of unique values in all columns to decide which ones to drop and put them in a list. 

col_drop = ["ID","Name","Nationality","Position","Club","Team & Contract","Joined",'Loan Date End','Contract','Hits'] 
df1 = df.drop(col_drop,axis=1)
df1.shape
```

```python
def convert_height(height):
    
    feet, inch = map(int, height.replace('"','').split("'"))
    inch = feet*12 + inch
    cm = inch * 2.54
    return cm
    
df1['Height'] = df1['Height'].apply(convert_height)    
df1['Height']
```

```python
df1['Weight']
df1['Weight'] = list(map(int, df1['Weight'].str.replace('lbs','')))
```

```python
df1['Weight'].dtypes
df1['Weight']
```

```python
def euro_check(monetary_value):
    for char in monetary_value:
        if char not in ('€', 'M', 'K'):
            return True
    return False

df['Value'].apply(euro_check).unique()
```

```python
def covert_value(value):
    if "M" in value:
        new_value = value.replace("€","").replace("M","")
        new_value = float(new_value)*1000000
    elif "K" in value:
        new_value = value.replace("€","").replace("K","")
        new_value = float(new_value)*1000
    else:
        new_value = value.replace("€","")
        new_value = float(new_value)

    return new_value

df1['Value'] = df1['Value'].apply(covert_value)
```

```python
df1['Wage'] = df1['Wage'].apply(covert_value)
display(df['Wage'].apply(euro_check).unique())
```

```python
df1['Release Clause'] = df1['Release Clause'].apply(covert_value)
display(df['Release Clause'].apply(euro_check).unique())
```

```python
def convert_position(nb):
    if isinstance(nb, int):
        return nb 
    
    if '-' in nb:
        score, bonus = map(int, nb.replace('-', '').split("+"))
        total = score - bonus
    else:
        score, bonus = map(int, nb.split("+"))
        total = score + bonus
    
    return total
```

```python
columns_to_convert = ['LS', 'ST', 'RS', 'LW', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM', 'RAM', 
                       'LM', 'LCM', 'CM', 'RCM', 'RM', 'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 
                       'LB', 'LCB', 'CB', 'RCB', 'RB', 'GK']

for column in columns_to_convert:
    df1[column] = df1[column].apply(convert_position)
```

```python
y = df1['OVA']
X = df1.drop(['OVA'], axis=1)
```

```python
X_num = X.select_dtypes(include = np.number)
X_cat = X.select_dtypes(include = object)
```

```python
from sklearn.preprocessing import MinMaxScaler 
from sklearn.preprocessing import StandardScaler

transformer = MinMaxScaler().fit(X_num)
x_standardized = transformer.transform(X_num)

X_num_scaled = pd.DataFrame(x_standardized, columns=X_num.columns)
X_num_scaled.head()
```

```python
from sklearn.preprocessing import OneHotEncoder  

encoder = OneHotEncoder(drop = "first")  #Creating encoder
encoder.fit(X_cat)  #fitting encoder with objects. 
X_cat_encoded = encoder.transform(X_cat).toarray()
X_cat_encoded = pd.DataFrame(X_cat_encoded, columns = encoder.get_feature_names_out())
X_cat_encoded
```

```python
df2 = pd.concat([X_cat_encoded, X_num_scaled], axis=1)
y = df1['OVA']
X = df2
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.40)  

lm = linear_model.LinearRegression() 
lm.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error, r2_score

# R2_Score
predictions = lm.predict(X_train)
predictions_test = lm.predict(X_test)

r2_score_train = r2_score(y_train, predictions)
r2_score_test = r2_score(y_test, predictions_test)

print("R2_Score for train: ", r2_score_train)  #r2_score shows how well my model can predict 
print("R2_Score for test: ", r2_score_test)


# MSE
mse_train = mean_squared_error(y_train, predictions)
mse_test = mean_squared_error(y_test, predictions_test)

print("MSE for train: ", mse_train)
print("MSE for test: ", mse_test)

# RMSE
rmse_train = np.sqrt(mean_squared_error(y_train,predictions))
rmse_test = np.sqrt(mean_squared_error(y_test,predictions_test))

print("RMSE for train: ", rmse_train)
print("RMSE for test: ", rmse_test)

# MAE
mae_train = mean_absolute_error(y_train, predictions)
mae_test = mean_absolute_error(y_test, predictions_test)

print("MAE for train: ", mae_train)
print("MAE for test: ", mae_test)
```
